```python
import re
import numpy as np
from pprint import pprint
import unicodedata

parteien = ['SPD', 'FDP', 'AfD', 'CDU/CSU', 'BÜNDNIS_90/DIE_GRÜNEN', 'DIE_LINKE']

# Füge zwei mal drei Spalten für jde Partei hinzu
for partei in parteien:
    for i in (1,2): 
        for field in ("Akteur","Aktion","Text"):
            df_comment[str(partei)+'_'+field+str(i)] = ""


# Ersetze Parteinamen bei denen Split zu Problemen führen würde
df_comment['Kommentar'] = df_comment['Kommentar'].replace(to_replace=r'((DIE )*LINKE)[N$:,]*', value='DIE_LINKE', regex=True)
df_comment['Kommentar'] = df_comment['Kommentar'].replace(to_replace=r'(BÜNDNIS(SES)*\s90\/DIE\sGRÜNEN)[$:,]*', value='BÜNDNIS_90/DIE_GRÜNEN', regex=True)

# Kommentare sind immer in Klammern, die entfernen wird hier. 
df_comment['Kommentar'] = df_comment['Kommentar'].replace(to_replace=r"(\(|\))", value='', regex=True)

#remove soft hyphens,
df_comment['Kommentar'] = df_comment['Kommentar'].replace(to_replace=r"\xad", value='', regex=True) 

# entferne Werte in [wert] die keine Parteien sind, manche MDBs haben eine Stadt in [] hinter ihrem Namen
df_comment['Kommentar'] = df_comment['Kommentar'].replace(to_replace=r'\[(?!(SPD|FDP|BÜNDNIS_90/DIE_GRÜNEN|CDU/CSU|DIE_LINKE|AfD))[\w\s\-]*\][\s]*', value='', regex=True)
        
df_comment
```

#### Vorbereitung der Regex Blöcke

```python
# Grundbausteine
typen = '(?:Heiterkeit und Beifall|Beifall|Lachen|Heiterkeit|Widerspruch|(Weitere )?Zurufe?|Gegenrufe?)'
parteien = '(?:SPD|FDP|AfD|CDU/CSU|BÜNDNIS_90\/DIE_GRÜNEN|DIE_LINKE|Bundesminister)'
akteure = '\s*(?:sowie|und)?[,\s]*(?:beim?|vo[nm])?\s*(?:de[rsmn])?\s*(?:Abgeordneten?|Abg\.)?\s*(?:de[rs]|vo[nm])?'
namen = '(?:[a-zA-ZÀ-ÖØ-öø-ſ ­\.\-]+)' # ­ ist ein weiches trennzeichen, wird z.B. bei Alexander Graf Lambsdorff der AfD genutzt- https://unicode-table.com/de/00AD/
texte = '(?:[a-zA-ZÀ-ÖØ-öø-ſ0-9 ­\.\?\!\,]+)'

# Blöcke zum verketten in einer großen Pattern
re_block={}
for BLOCK in ['A']:                    # Mehr as Block A nicht notwendig, weil Kommentare an "-" getrennt werden 
    bl={}
    for AKT in [1,2,3,4,5,6]:
        #für jeden Akteur
        bl[AKT] = {}
        inst = BLOCK+str(AKT)
        # helper blocks, werden in anderen blocks verschachtelt
        #bl[AKT]["partei_mdb"] = "(?P<partei_mdb_"+inst+">"+parteien+")"
        bl[AKT]["partei"]     = "(?P<partei_"+inst+">"+parteien+")"
        bl[AKT]["text"]       = "(?::\s*)(?P<text_"+inst+">"+texte+")"
        bl[AKT]["mdb"]        = "(?P<mdb_"+inst+">"+namen+")?[\,\s]*(?:\[?(?P<partei_mdb_"+inst+">"+parteien+")\]?)"

        #blocks zur nutzung in patterns
        bl[AKT]["typ"] = "(?P<typ_"+inst+">"+typen+")"
        bl[AKT]["akteur"] = "(?:\
(?P<akteur_"+inst+"_p>"+akteure+")\s*"+bl[AKT]["partei"]+"|\
(?P<akteur_"+inst+"_m>"+akteure+")\s*"+bl[AKT]["mdb"]+")(?:"+bl[AKT]['text']+")?" 

    re_block[BLOCK] = bl
    
pprint(re_block)
```

## Auslesen der Kommentare
##### 1.) Erstellen der Pattern aus Blöcken
##### 2.) Zuweisung der gematchten Werte in Spalten

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_colwidth', 500)
pd.set_option('display.width', 1000)

####### Erstelle Pattern für Match ########

pattern_from_blocks = ""

for block in re_block.values(): # aktuell 2020-07-16 nur "A"
    AKT=1
    
    single_pattern_from_blocks = '(?:'+block[AKT]['typ']+block[AKT]['akteur']+')?'
    while AKT < len(block.keys()):                               
        AKT = AKT + 1
        single_pattern_from_blocks += '(?:'+block[AKT]['akteur']+')?' # Füge weitere Akteure hinzu (die z.B auch Beifall klatschen)
    pattern_from_blocks += single_pattern_from_blocks
    # +"(?:[–\s]*)?"   # Nächste Aktion im selben Kommentar (z.B.Zuruf)
                       # nicht nötig, da wir Kommentare eh an "-" auftrennen und somit nur eine Aktion pro Kommentar haben
    
used_pattern = pattern_from_blocks                                    # used_pattern ist das fertig erstellte Pattern, das alle Bestandteile eines Kommentars erkennen sollte
#print(used_pattern)
# print("###############\n")



######## Auslesen der Kommentare ####### 
from pprint import pprint
from collections import defaultdict  # available in Python 2.5 and newer
 # used for comments_by_party, based on https://stackoverflow.com/questions/1692388/python-list-of-dict-if-exists-increment-a-dict-value-if-not-append-a-new-dict
import unicodedata
from datetime import datetime

parsing_issues = 0
for i in range(0, len(df_comment['Kommentar'])):
    if i % 3000 == 0: # every 3000th comment, give some status information to see progress, should be around every second
        print("%s: processing comment %d of %d" % (datetime.now().strftime("%H:%M:%S"),i,len(df_comment['Kommentar'])))
    comment = unicodedata.normalize("NFKC", df_comment['Kommentar'][i])     # normalisiere jedes Kommentar
                                                                            # https://docs.python.org/2/library/unicodedata.html#unicodedata.normalize
                                                                            # zb um aus \xa0 dem non-breaking space zwischen "Dr. Vorname" ein richtiges leerzeichen zu machen    
    if comment != "NaN":                                                    # ignoriere NaNs
        comments_by_party = defaultdict(int)                                # comments_by_party als counter für jede Partei -> zur befüllung von Aktion1 oder Aktion2
        for single_comment in re.split("\s–\s",comment):                    # teile Kommentare auf die mit "-" getrennt sind
            matches = re.match(used_pattern, single_comment)
            
            if not matches:
                print("Found no matches for comment: %s" % single_comment)  
            elif matches:
                
                match_dict = {k: v.strip() for k, v in matches.groupdict().items() if v is not '' and v is not None}
                 # copied from https://stackoverflow.com/questions/12118695/efficient-way-to-remove-keys-with-empty-strings-from-a-dict
                 # added .strip() to remove whitespace around text and akteur matches
                
                if (len(match_dict) < 2):
                    parsing_issues += 1                       # ??? Wenn weniger als 2 matches muss ein Fehler vorliegen
                    continue
                BLOCK="A"
                c_groups={} # gruppe der zusammengehörigen matches ??? wofür ist dieses dic gut?
                c_groups['success'] = {}
                c_groups['failures'] = {}
                c_groups['matches'] = match_dict
                c_groups['comment_line'] = single_comment
                c_groups['full_comment'] = comment
                #(c_groups)

                for AKT in [1,2,3,4,5,6]:
                    c = {} # dict das immer nur die zusammengehörigen Matches fasst 
                           # um sie dann gemeinsam in df_comment zu schreiben
                    inst = BLOCK+str(AKT)
                    #print("Looking at match groups with '%s'" % (inst))
                    
                    # da der MdB spezifischer ist, prüfen wir diesen zuerst
                    if 'mdb_'+inst in match_dict:
                        c['Akteur'] = match_dict["mdb_"+inst]
                        c['party']  = match_dict["partei_mdb_"+inst]
                    elif 'akteur_'+inst+"_p" in match_dict:
                        c['Akteur'] = match_dict['akteur_'+inst+"_p"]
                        if "Abg" in c['Akteur']:
                            c['Akteur'] = "teile"
                        else:
                            c['Akteur'] = "ganze"
                        c['party']  = match_dict["partei_"+inst]
                    else:
                        c['failure_reason'] = "akteur nicht erkannt - grund unklar"
                        
                    if 'text_'+inst in match_dict:
                        c['Text']   = match_dict["text_"+inst]
                        c['Aktion'] = "Zuruf"
                            
                    elif 'typ_'+BLOCK+"1" in match_dict: # typ gibt es pro kommentar nur einmal, also auch nur eine matchgroup typ_A1, kein A2
                        c['Aktion'] = match_dict['typ_'+BLOCK+"1"]
                    else:
                        c['failure_reason'] = "weder text noch typ gefunden, unklar was hier geparsed wurde!!"

                    
                    if not "failure_reason" in c:
                        #es wurde erfolgreich geparsed, dann machen wir jetzt noch etwas Datenbereinigung
                        # A + B haben dabe nichts miteinander zu tun
                        #A
                        if c['Akteur'] == "ganze" and c['Aktion'] in ("Zuruf","Zurufe","Widerspruch","Gegenruf","Gegenrufe"):
                                #Anonyme Zurufe werden oben sonst der ganzen Partei zugeschrieben
                                # das ist bei wörtlichen Zurufen unsinn, also überschreiben wir das hier
                                c['Akteur'] = "teile"
                        #B
                        if comments_by_party[c['party']] >= 2:
                            c['failure_reason'] = "overflow - nur die ersten 2 erfolgreich geparsten Beiträge einer Partei werden erfasst\
                            Ein Feld für <Partei>3 gibt es im df_comment eh nicht." 
                        else:
                            comments_by_party[c['party']] += 1 
                            # zähle die zahl der kommentare pro Partei hoch, 
                            # startet dank defaultdict bei 0 wenn es bisher kein Kommentar gab
                    if "failure_reason" in c:
                      c_groups['failures'][inst] = c
                    else:
                      c_groups['success'][inst] = c
                    
                if c_groups['success']:
                    #if single_comment=="Beifall bei der CDU/CSU, der SPD, der FDP, der DIE_LINKE und dem BÜNDNIS_90/DIE_GRÜNEN":
                      #print(df_comment.loc[i,"Rede_Text"])
                      for inst in c_groups['success']:
                        #print("_____DEBUG_____\nFound a match for %s" % (inst))
                        c = c_groups['success'][inst]
                        #pprint(c)
                        for field in ("Aktion","Akteur","Text"):
                            if field in c:
                                df_comment.loc[i, c['party']+"_"+field+str(comments_by_party[c['party']])] = c[field]
                else:
                  if "debug" == "no_debug":
                    accepted_failure = False
                    for inst in c_groups['failures']:
                        c = c_groups['failures'][inst]
                        if c['failure_reason'].startswith("overflow -"):
                            accepted_failure = True
                    if not accepted_failure:
                        print("##############################Parsing Failure!\n")
                        pprint(c_groups)

                    
            

print("%s: processed all comments" % (datetime.now().strftime("%H:%M:%S")))
```
