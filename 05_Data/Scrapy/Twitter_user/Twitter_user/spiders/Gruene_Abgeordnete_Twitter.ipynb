```python
import json

class JsonWriterPipeline(object):

    def open_spider(self, spider):
        self.file = open('Gruene_twitter.json', 'w')

    def close_spider(self, spider):
        self.file.close()

    def process_item(self, item, spider):
        line = json.dumps(dict(item)) + "\n"
        self.file.write(line)
        return item
```

```python
import scrapy
from scrapy.crawler import CrawlerProcess
import logging
from scrapy.utils.log import configure_logging
from urllib.parse import urlparse



configure_logging(install_root_handler=False)
logging.basicConfig(
    filename='Gruene_log.txt',
    filemode = 'w',
    format='%(levelname)s: %(message)s',
    level=logging.DEBUG
)



class GrueneSpider(scrapy.Spider):
    name = "Gruene_Abgeordnete_Twitter"
    custom_settings = {
        'LOG_LEVEL': logging.INFO,
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1}, # Used for pipeline 1
        'FEED_FORMAT':'json',                                 # Used for pipeline 2
        'FEED_URI': 'Gruene_twitter.json'                     # Used for pipeline 2
    }

    def start_requests(self):
        urls = [
            'https://www.gruene-bundestag.de/abgeordnete',
        ]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        #for href in response.css("div.teaserList.teaserType-weitereInfoTeaser.teaserList--tiny a::attr('href')"):
        for link in response.css("article.abgeordneteTeaser a::attr(href)").extract():
            #print(link)
            #url = response.urljoin(link)
            #url = urlparse.urljoin(response.url, link)
            yield response.follow(link, callback=self.parse_dir_contents)

#            print(url)
#            yield scrapy.Request(link, callback = self.parse_dir_contents)

    def parse_dir_contents(self, response):
        for delegate in response.css('div.documentContent__sharingContainer'):
            print(delegate)
            yield {
                #'text': delegate.css('li.twitter::text').getall(),
                'name': delegate.xpath('//a[contains(@href, "twitter.com")]/@href').getall(),
            }
```

```python
process3 = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process3.crawl(GrueneSpider)
process3.start()
```
